```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
data = pd.read_csv('dataout0.csv', header=None)
x = data[0].values
y = data[1].values
inliers = data[2].values
labels = data[3].values
X = np.array(list(zip(x, y)))

centroids = pd.read_csv('centroids_dataout0.csv', header=None)
c_x = centroids[0].values
c_y = centroids[1].values
ctrs = np.array(list(zip(c_x, c_y)))
```

```python
fig = plt.figure(figsize=(16, 9))
ax = plt.subplot(1, 1, 1)
k = 9
for i in range(k):
    points = np.array([X[j] for j in range(len(X)) if labels[j] == i])
    ax.scatter(points[:, 0], points[:, 1], cmap='rainbow')
    ax.scatter(ctrs[i, 0], ctrs[i, 1], marker='*', s=100, c='#050505')
for idx, val in enumerate(inliers):
    if(not(val)):
        ax.scatter(X[idx, 0], X[idx, 1], marker='x', s=100, c='#555555')
```
