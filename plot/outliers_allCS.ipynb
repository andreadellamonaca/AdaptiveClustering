```python
import numpy as np
import os
import pandas as pd
from matplotlib import pyplot as plt
```

```python
directory = './iris/'
for i in range(1):
    data = pd.read_csv(directory+'dataout'+str(i)+'.csv', header=None)
    x = data[0].values
    y = data[1].values
    inliers = data[2].values
    labels = data[3].values
    X = np.array(list(zip(x, y)))

    centroids = pd.read_csv(directory+'centroids_dataout'+str(i)+'.csv', header=None)
    c_x = centroids[0].values
    c_y = centroids[1].values
    ctrs = np.array(list(zip(c_x, c_y)))
    
    fig = plt.figure(figsize=(16, 9))
    ax = plt.subplot(1, 1, 1)
    k = len(c_x)
    for q in range(k):
        points = np.array([X[j] for j in range(len(X)) if labels[j] == q])
        ax.scatter(points[:, 0], points[:, 1], cmap='rainbow')
        ax.scatter(ctrs[q, 0], ctrs[q, 1], marker='*', s=50, c='#050505')
    for idx, val in enumerate(inliers):
        if(not(val)):
            ax.scatter(X[idx, 0], X[idx, 1], marker='x', s=100, c='#555555')
    #plt.savefig(directory+str(i)+'.png')
```
